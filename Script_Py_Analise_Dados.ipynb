```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)

y = np.random.randint(low = 1, high = 1500, size = 10)
```

```python
y
```

```python
plt.plot(y)
plt.title('Matplotlib')
plt.xlabel('Largura')
plt.ylabel('Altura')
```

```python
# prompt: plote um grafico cujo o y tem a cor red marker 'o' ms = 5 mec = 'k' markerfacecolor = 'w' ls = '-.'

plt.plot(y, color = 'red', marker = 'o', ms = 5, mec = 'k', markerfacecolor = 'g', ls = '-.')
plt.title('Matplotlib')
plt.xlabel('Largura')
plt.ylabel('Altura')
```

```python
plt.plot(y*2, marker = '+', color = 'g', ms = 5)
plt.plot(y, color = 'red', marker = 'o', ms = 5, mec = 'k', markerfacecolor = 'g', ls = '-.')
```

```python
x = np.random.normal(10, 0.5, 100)
y = np.random.normal(0, 20, 100)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.scatter(x, y, marker = 'o', color = 'red', label = 'data 1', alpha = 0.5)
ax.scatter(x*0.5, y*0.5, marker = 'o', color = 'black', label = 'data 2', alpha = 0.7)
ax.legend()
```
