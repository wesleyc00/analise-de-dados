```python
# Bibliotecas

import pandas as pd
import plotly.express as px
```

```python
# Criar cidades aleatorias
# Cidades

cidades = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre', 'Curitiba','Moju','Abaetetuba','Barcarena']

Estados = ['SP', 'RJ', 'MG', 'RS', 'PR','PA','PA','PA']

# Gerando a Latitude
Latitude = [-23.6666, -23.6944, -23.6865, -23.6226, -23.6687,
            -23.7141, -23.7452, -23.5489]

# Gerando a Longitude
Longitude = [-46.5322, -46.5654, -46.6234, -46.5489, -46.4614,
             -46.4137, -46.4022, -46.6388]

# Valor
Vendas = [100, 120, 90, 50, 70, 90, 250, 400]
```

```python
# Dicionario

dados = {'Cidade': cidades, 'Estado': Estados, 'Latitude': Latitude, 'Longitude': Longitude, 'Vendas': Vendas}

# Base de dados

Base_Dados= pd.DataFrame(dados)

Base_Dados
```

```python
# Mapa de CAlor Geografico

px.density_mapbox(
    Base_Dados,
    lat='Latitude',
    lon='Longitude',
    z='Vendas',
    radius=30,
    center=dict(lat=-23.700, lon=-46.5555),
    zoom=7,
    mapbox_style='stamen-terrain'
)
```
