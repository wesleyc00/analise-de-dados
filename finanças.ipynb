```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definir tema
sns.set_theme(style = 'darkgrid')
```

```python
# Dados
Base_dados = sns.load_dataset('tips')

# Verifica
Base_dados.head()
```

```python
# Renomeando colunas
Base_dados.rename(columns = {'total_bill': 'Conta_total',
                             'tip': 'Gorjeta',
                             'sex': 'Sexo',
                             'smoker': 'Fumante',
                             'day': 'Dia',
                             'time': 'Tempo',
                             'size': 'Tamanho'
                            }, inplace = True)
```

```python
Base_dados.head()
```

```python
# Grafico relplot
sns.relplot(data = Base_dados, x = 'Conta_total', y = 'Gorjeta')
```

```python
# Grafico relplot
# Passando outro parametro como classe no grafico
sns.relplot(data = Base_dados, x = 'Conta_total', y = 'Gorjeta', hue = 'Sexo')
```

```python
# Grafico relplot
sns.relplot(data = Base_dados, x = 'Conta_total', y = 'Gorjeta', kind = 'line')
```

```python
# Grafico relplot
# Fumantes
# Gráfico de linha
sns.relplot(data = Base_dados, x = 'Conta_total', y = 'Gorjeta', hue = 'Fumante', kind='line')
```

```python
# Grafico relplot
sns.histplot(data = Base_dados, x = 'Conta_total')
```

```python
sns.histplot(data = Base_dados, x = 'Conta_total', hue = 'Fumante')
```

```python
sns.barplot(data = Base_dados, x = 'Sexo', y = 'Conta_total', hue = 'Fumante')
```

```python
sns.pairplot(data = Base_dados)
#
```

```python
sns.pairplot(data = Base_dados, hue = 'Sexo')
```

```python
sns.pairplot(data = Base_dados, kind = 'kde')
```

```python
sns.boxplot(data = Base_dados, x = 'Dia', y = 'Conta_total')
```
